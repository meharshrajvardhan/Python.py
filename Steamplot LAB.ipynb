# Stem PLOT

```python
import matplotlib.pyplot as plt

x = [2,4,5,7,8,9,1,3,6]
y = [1,3,2,9,6,8,7,6,4]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=5,basefmt="g",label="python",orientation="vertical",)
plt.legend()
plt.show()
```

# BOX PLOT

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
x1 =[10,20,40,50,30,60,90]

y = [x,x1]

plt.boxplot(y,labels=["python","c++"],showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r"))

             
plt.show()                     
```

# Stack Plot

```python
import matplotlib.pyplot as plt

x = [1,5,7,9,11,15,17,9,20]
area = [1,7,6,14,16,15,4,7,9]
plot = [2,7,10,7,16,8,4,7,9]
squre = [4,5,7,9,7,15,27,18,14]

l= ["area","plot","squre"]
plt.stackplot(x,area,plot,squre,labels=l,colors=["r","g","m"])
plt.title("Land")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

# Step Plot

```python
import matplotlib.pyplot as plt

x = [1,5,7,9,11,15,17,9,20]
y = [1,5,7,9,11,15,17,9,20]

plt.step(x,y,marker=("o"),ms=10,mfc="g",label="python")
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend(loc=2)
plt.show()
```
