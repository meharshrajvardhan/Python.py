Pie Plot

```python
import matplotlib.pyplot as plt

x= [10,20,30,40]
y= ["c","c++","java","python"]
ex= [0.3,0.0,0.0,0.0]
c= ["r","b","g","m"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=1,startangle=0,textprops={"fontsize":15}
       ,wedgeprops={"linewidth":5},rotatelabels=True,counterclock=False)
plt.title("Dashboard",fontsize=14)
plt.legend(loc=1)
plt.show()
```
